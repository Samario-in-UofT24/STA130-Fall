```python
import pandas as pd
import numpy as np

# Creating a sample DataFrame with missing values
data = {
    'CustomerID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', np.nan, 'Eve'],
    'Age': [25, np.nan, 30, 22, 29],
    'City': ['New York', 'Los Angeles', np.nan, 'Chicago', 'Houston'],
    'IrrelevantColumn': [np.nan, np.nan, np.nan, np.nan, np.nan]  # Irrelevant column with all NaNs
}
df5 = pd.DataFrame(data)
df5
```

```python
del df5['IrrelevantColumn']
df5
```

```python
df5.dropna()
```
